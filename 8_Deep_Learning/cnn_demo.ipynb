# A demo on the principles of CNNs

First, let's load a picture of a cat. 

```python
#Checking out a color image
from scipy import ndimage as ndi
import numpy 
from skimage import data
import matplotlib.pyplot as plt
import skimage


cat = data.chelsea()
print(f'Shape: {cat.shape}')
print("Values min/max", cat.min(), cat.max())
plt.imshow(cat)
```

```python
graycat = skimage.color.rgb2gray(cat)
print (graycat.shape)
plt.imshow(graycat, cmap=plt.cm.gray)
```

```python

line_kernel = numpy.array([[-1,-1,-1],[2,2,2],[-1,-1,-1]])
conved = ndi.convolve(graycat, line_kernel)
plt.imshow(conved)
```

```python
line_kernel = numpy.array([[-1,-1,-1],[2,2,2],[-1,-1,-1]])
conved = ndi.convolve(graycat, line_kernel.transpose())
plt.imshow(conved)
```
